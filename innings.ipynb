```python
import pandas as pd
import altair as alt
```

```python
alt.themes.enable('opaque')
alt.data_transformers.enable('json')
```

```python
df = pd.read_csv("./input/war_daily_pitch.txt")
```

```python
alt.Chart(df).mark_bar().encode(
    x=alt.X("year_ID:O", axis=alt.Axis(title="Year", tickCount=10)),
    y=alt.Y("median(salary):Q")
).properties(width=800).transform_filter(alt.datum.year_ID > 1920)
```
